```python
x = ['Hubert:a1a2', 'Robin:a1a2']
```

```python
names = [xx.split(':')[0] for xx in x]
```

```python
moves = [xx.split(':')[1] for xx in x]
```

```python
name_to_move = dict(zip(names, moves))
```

```python
list(name_to_move.values())[0]
```

```python
with open('toto', 'w') as f:
    for n in name_to_move:
        f.write(f'{n}:{name_to_move[n]}\n')
```
